# Perturbation series

Beware of the techniques available

## Three steps

1. Convert the equation into perurbation form ( introduce small parameters $\epsilon$)
2. Write the Solutions as a perturbation series and calculate the series coefficients
3. Reconstruct the solution by summing the perurbation series

## How does boundary conditions fit into the equation

epsilon does not appear in boundary conditions thus 

<b>All Boundary conditions map onto first term (Zero Order)</b>

thus $ A + Bx = 1 + x$


Boundary conditions for other terms are zero 

think of boundary conditions having epsilon terms of coefficient 0





Only consider first few terms depending on the size of epsilon.

This is the perturbation solution to the equation.


This is not a taylor series expansion -> this is a point solution ( local)

Perturbation series are valid over the entire solution state ( global)
